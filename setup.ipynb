## Dataset installation

We are going to download the dataset from Hugging Face from the link:

[https://huggingface.co/datasets/ErfanMoosaviMonazzah/fake-news-detection-dataset-English](https://huggingface.co/datasets/ErfanMoosaviMonazzah/fake-news-detection-dataset-English)

We're saving the 3 splits (train, validation and test) in the directory 'Dataset'

```python
import pandas as pd

splits = {'train': 'train.tsv', 'validation': 'validation.tsv', 'test': 'test.tsv'}

for split in splits:
    df = pd.read_csv(f"hf://datasets/ErfanMoosaviMonazzah/fake-news-detection-dataset-English/{splits[split]}", sep="\t")
    df.to_csv(f"Dataset/{split}.csv", index=False)
```
